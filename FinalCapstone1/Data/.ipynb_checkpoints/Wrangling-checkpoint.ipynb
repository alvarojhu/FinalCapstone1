```python
import pandas as pd
```

```python
stats2018 = pd.read_csv('2017-2018stats.txt')
salaries2019 = pd.read_csv('2018-2019salaries.txt', index_col = 0)
```

```python
stats2018[['Name','string']] = stats2018['Player'].str.split('\\', expand = True)
stats2018 = stats2018.drop(['Player'],1)
stats2018 = stats2018[['Name','Pos','Tm','G','GS', 'FG', 'FG%', '3P', '3P%', 'eFG%', 'FT%', 'ORB','TRB','AST','STL','BLK','PS/G']]
stats2018.head()
```

```python
#multiples = stats2018.loc[stats2018['Tm']=='TOT']
stats2018u = stats2018.drop_duplicates(subset = ['Name']).set_index(['Name'])
stats2018u.head()

    

```

```python
salaries2019[['Name','string']] = salaries2019['Player'].str.split('\\', expand = True)
salaries2019 = salaries2019.drop(['Player'],1)
salaries2019 = salaries2019.drop(['string'],1).drop_duplicates(subset = ['Name']).set_index(['Name'])
salaries2019.head()
```

```python
data2018 = pd.concat([salaries2019,stats2018u], axis = 1)
highest = data2018.sort_values(by = '2018-19', ascending = False).dropna(subset = ['2018-19', 'G'])
highest
```
